```python
import pandas as pd
import numpy as np

df_abcd = pd.read_csv("LesionParameters.csv")
df_gt = pd.read_csv("ISIC2/ISIC_2019_Training_GroundTruth.csv")

df_abcd['image'] = df_abcd['image'].str.replace('.jpg', '', regex=False).str.strip()

df_gt['image'] = df_gt['image'].str.strip()
df_gt['image'] = df_gt['image'].str.replace('_downsampled', '', regex=False)

print(f"Ilość obrazów w Twoim zbiorze ABCD przed złączeniem: {len(df_abcd)}")

df_final = pd.merge(df_abcd, df_gt, on='image', how='inner')

class_names = ['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC', 'UNK']
df_final['label'] = df_final[class_names].to_numpy().argmax(axis=1)

df_final = df_final.drop(columns=class_names)

print(f"Ilość obrazów PO złączeniu: {len(df_final)}")
print(df_final.head())
```
