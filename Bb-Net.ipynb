```python
from ultralytics import YOLO

model = YOLO('yolo11m.pt')
def main():
    results = model.train(
        data='/mnt/454a7ba1-a082-4c2e-8228-781fbc12fe3b/Projekty/Studia/SieciAng/Projekt/data/data.yaml',
        epochs=100,
        patience=30,
        imgsz=640,
        device=0,
        workers=4,
    )
```

```python
main()
```

```python
metrics = model.val(data='data/data.yaml', split='test')

f1_score = metrics.box.f1.mean() 
precision = metrics.box.p.mean()
recall = metrics.box.r.mean()
map50 = metrics.box.map50
map95 = metrics.box.map

print(f"--- Statystyki dla klasy: {metrics.names[0]} ---")
print(f"Precision (Precyzja): {precision:.4f}")
print(f"Recall (Czułość):     {recall:.4f}")
print(f"F1-Score:             {f1_score:.4f}")
print(f"mAP50:                {map50:.4f}")
print(f"mAP50-95:             {map95:.4f}")
```
